```python
import matplotlib.pyplot as plt
import numpy as np
#import random
%matplotlib notebook
```

```python
#interwoven 3D grid with index 2*1*1, every copy of the motif in a different color:

num_steps = 25 #choosing how densely the lines are sampled. Namely, each straight line is sampled by num_step points.

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')



#These for-loops will copy the motif in 2x2x2 different translates.
for x_iter in range(0,2,1):
    for y_iter in range(0,2,1):
        for z_iter in range(0,2,1):
            
            #I want to make a path from (0,0,0) to (2,0,0) thus in x-direction, but skipping one integer point.
            #To avoid intersection with its copies, I have to make the path curved. 
            #Thus it consits of 3 straight lines: first part, second part, and third part, together forming a curve.
            #The second and third part form an arc, under which the copy of the first part can pass underneath.
            x_direction_first_part_x_vals = np.linspace(x_iter,      x_iter+0.3, num_steps) 
            x_direction_first_part_y_vals = np.linspace(y_iter,      y_iter+0.4, num_steps)
            x_direction_first_part_z_vals = np.linspace(z_iter,      z_iter,     num_steps)

            x_direction_second_part_x_vals = np.linspace(x_iter+0.3, x_iter+1.2, num_steps) 
            x_direction_second_part_y_vals = np.linspace(y_iter+0.4, y_iter+0.2, num_steps)
            x_direction_second_part_z_vals = np.linspace(z_iter,     z_iter+0.4, num_steps)
            
            x_direction_third_part_x_vals = np.linspace(x_iter+1.2,  x_iter+2,   num_steps) 
            x_direction_third_part_y_vals = np.linspace(y_iter+0.2,  y_iter,     num_steps)
            x_direction_third_part_z_vals = np.linspace(z_iter+0.4,  z_iter,     num_steps)
            
            
            #In y-direction I decided not to skip any integer points, so connecting (0,0,0) directly with (0,1,0).
            #This can be done with a straight line, thus only one part, without risking intersections.
            y_direction_first_part_x_vals = np.linspace(x_iter,      x_iter,     num_steps) 
            y_direction_first_part_y_vals = np.linspace(y_iter,      y_iter+1,   num_steps)
            y_direction_first_part_z_vals = np.linspace(z_iter,      z_iter,     num_steps)

            #Same in z-direction.
            z_direction_first_part_x_vals = np.linspace(x_iter,      x_iter,     num_steps) 
            z_direction_first_part_y_vals = np.linspace(y_iter,      y_iter,     num_steps)
            z_direction_first_part_z_vals = np.linspace(z_iter,      z_iter+1,   num_steps)

            
            
            #Concatenating the individual pieces:
            x_vals = np.concatenate((x_direction_first_part_x_vals, x_direction_second_part_x_vals, x_direction_third_part_x_vals, y_direction_first_part_x_vals, z_direction_first_part_x_vals))
            y_vals = np.concatenate((x_direction_first_part_y_vals, x_direction_second_part_y_vals, x_direction_third_part_y_vals, y_direction_first_part_y_vals, z_direction_first_part_y_vals))
            z_vals = np.concatenate((x_direction_first_part_z_vals, x_direction_second_part_z_vals, x_direction_third_part_z_vals, y_direction_first_part_z_vals, z_direction_first_part_z_vals))

            #scatter plot for each possible shift from the for-loop in a different color.
            ax.scatter(x_vals, y_vals, z_vals)

plt.show()
```

```python
#interwoven 3D grid with index 2*1*1, the two components in two different colors:

num_steps = 25

fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(projection='3d')

#These for-loops will copy the motif in 2x2x2 different translates, but skipping one in x-direction, so that we only get the connected piece.
for x_iter in range(0,4,2):
    for y_iter in range(0,2,1):
        for z_iter in range(0,2,1):

            x_direction_first_part_x_vals = np.linspace(x_iter,      x_iter+0.3, num_steps) 
            x_direction_first_part_y_vals = np.linspace(y_iter,      y_iter+0.4, num_steps)
            x_direction_first_part_z_vals = np.linspace(z_iter,      z_iter,     num_steps)

            x_direction_second_part_x_vals = np.linspace(x_iter+0.3, x_iter+1.2, num_steps) 
            x_direction_second_part_y_vals = np.linspace(y_iter+0.4, y_iter+0.2, num_steps)
            x_direction_second_part_z_vals = np.linspace(z_iter,     z_iter+0.4, num_steps)
            
            x_direction_third_part_x_vals = np.linspace(x_iter+1.2,  x_iter+2,   num_steps) 
            x_direction_third_part_y_vals = np.linspace(y_iter+0.2,  y_iter,     num_steps)
            x_direction_third_part_z_vals = np.linspace(z_iter+0.4,  z_iter,     num_steps)
            
            
            
            y_direction_first_part_x_vals = np.linspace(x_iter,      x_iter,     num_steps) 
            y_direction_first_part_y_vals = np.linspace(y_iter,      y_iter+1,   num_steps)
            y_direction_first_part_z_vals = np.linspace(z_iter,      z_iter,     num_steps)

            
            z_direction_first_part_x_vals = np.linspace(x_iter,      x_iter,     num_steps) 
            z_direction_first_part_y_vals = np.linspace(y_iter,      y_iter,     num_steps)
            z_direction_first_part_z_vals = np.linspace(z_iter,      z_iter+1,   num_steps)

            
            

            x_vals = np.concatenate((x_direction_first_part_x_vals, x_direction_second_part_x_vals, x_direction_third_part_x_vals, y_direction_first_part_x_vals, z_direction_first_part_x_vals))
            y_vals = np.concatenate((x_direction_first_part_y_vals, x_direction_second_part_y_vals, x_direction_third_part_y_vals, y_direction_first_part_y_vals, z_direction_first_part_y_vals))
            z_vals = np.concatenate((x_direction_first_part_z_vals, x_direction_second_part_z_vals, x_direction_third_part_z_vals, y_direction_first_part_z_vals, z_direction_first_part_z_vals))

            ax.scatter(x_vals, y_vals, z_vals, color='k') #plotting one connected component in black
            ax.scatter(x_vals+1, y_vals, z_vals, color='b') #and the other, shifted by (1,0,0) in blue

plt.show()
```
